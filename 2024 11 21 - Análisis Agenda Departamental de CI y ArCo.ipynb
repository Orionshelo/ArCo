```python
import nltk
from nltk.corpus import stopwords


# Descargar lista de palabras vacías en español
nltk.download('stopwords')
spanish_stopwords = stopwords.words('spanish')
```

# Intersección 

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity


# Extraer las columnas relevantes
#alcances = agenda_nacional_ci['Alcances'].dropna().astype(str)
#eje_comp_inno = agenda_nacional_ci['Eje para la Competitividad e Innovación'].dropna().astype(str)
objetivo_general_ppi = agenda_departamental_ci['Objetivo General'].dropna().astype(str)
#departamento_ppi = agenda_departamental_ci['Departamento'].dropna().astype(str)
descripciones_largas = arco['Descripción'].dropna().astype(str)
nombres_instrumentos = arco['Nombre del Instrumento'].dropna().astype(str)
entidad_instrumentos = arco['Entidad'].dropna().astype(str)


# Vectorizar los textos usando TF-IDF con las palabras vacías en español
vectorizer = TfidfVectorizer(stop_words=spanish_stopwords)
tfidf_matrix = vectorizer.fit_transform(pd.concat(
[
#alcances, eje_comp_inno, 
objetivo_general_ppi, descripciones_largas
], ignore_index=True))

# Calcular la similitud de coseno entre "Acciones/Proyectos" y "Descripción larga"
similarity_matrix = cosine_similarity(tfidf_matrix[:len(objetivo_general_ppi)], tfidf_matrix[len(objetivo_general_ppi):])

print(len(nombres_instrumentos))
print(similarity_matrix.shape)


# Crear un DataFrame para almacenar resultados
resultados_similitud_interseccion = pd.DataFrame({
    #'Alcances': alcances.values,
    #'Eje para la Competitividad e Innovación': eje_comp_inno.values,
    'Objetivo General PPI más similar': objetivo_general_ppi.values,
    #'Departamento PPI' : departamento_ppi,
    'Nombre del Instrumento más similar': [nombres_instrumentos.iloc[similarity.argmax()] for similarity in similarity_matrix],
    'Entidad asociada al instrumento': [entidad_instrumentos.iloc[similarity.argmax()] for similarity in similarity_matrix],
    'Descripción larga más similar': [descripciones_largas.iloc[similarity.argmax()] for similarity in similarity_matrix],
    'Similitud': [similarity.max() for similarity in similarity_matrix],
    'Tiene similitud': ['Sí' if similarity.max() > 0.2 else 'No' for similarity in similarity_matrix]  # Umbral ajustable
})



# Mostrar resultados
print(resultados_similitud_interseccion.head())
```
