```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from langchain.chains import RetrievalQA
from langchain_groq import ChatGroq

import getpass
import os
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# Configurar la API de Groq
if not os.environ.get("GROQ_API_KEY"):
    os.environ["GROQ_API_KEY"] = getpass.getpass("Enter API key for Groq: ")

# Inicializar el modelo de Groq
model = ChatGroq(model="llama-3.3-70b-versatile")

# Cargar los datos
nombre_archivo = "2024 11 01 - Base Original.xlsx"
data_arco = pd.read_excel(nombre_archivo)

# Verificar columnas relevantes
data_arco = data_arco[['Nombre del Instrumento', 'Descripción']]
data_arco = data_arco.dropna()

# Crear embeddings para las descripciones
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
vector_store = FAISS.from_texts(data_arco['Descripción'].tolist(), embeddings)

# Crear el sistema de preguntas y respuestas (Q&A)
qa_chain = RetrievalQA.from_chain_type(
    llm=model,
    retriever=vector_store.as_retriever()
)


#def recomendar_instrumentos(query):
#    """
#    Recomienda instrumentos de internacionalización basados en una consulta específica.
#    
#    Parameters:
#        query (str): Consulta del usuario describiendo el instrumento deseado.
#        
#    Returns:
#        dict: Instrumentos recomendados con sus descripciones.
#    """
#    resultados = qa_chain.run(query)
#    return resultados

def recomendar_instrumentos(query):
    """
    Recomienda instrumentos de financiamiento basados en una consulta específica.
    
    Parameters:
        query (str): Consulta del usuario describiendo el instrumento deseado.
        
    Returns:
        list: Lista de diccionarios con el nombre y descripción de los instrumentos recomendados.
    """
    resultados = qa_chain.run(query)
    instrumentos_recomendados = []

    # Crear un modelo TF-IDF para medir similitud
    tfidf = TfidfVectorizer()
    tfidf_matrix = tfidf.fit_transform(data_arco['Descripción'])

    for descripcion in resultados.split("\n"):
        descripcion = descripcion.strip()
        if descripcion:
            query_vector = tfidf.transform([descripcion])
            similitudes = cosine_similarity(query_vector, tfidf_matrix)
            indices_similares = similitudes.argsort()[0][-5:][::-1]  # Obtener los 5 más similares

            for idx in indices_similares:
                instrumento = {
                    "Nombre del Instrumento": data_arco.iloc[idx]['Nombre del Instrumento'],
                    "Descripción": data_arco.iloc[idx]['Descripción']
                }
                # Generar una explicación usando el modelo
                explicacion = model.invoke(
                    f"Explica por qué el instrumento '{instrumento['Nombre del Instrumento']}' es adecuado para la consulta: {query}"
                )
                instrumento["Explicacion"] = explicacion
                instrumentos_recomendados.append(instrumento)
    return instrumentos_recomendados


# Ejemplo de uso
consulta_usuario = "Dame una ruta de instrumentos para que un usuario acceda de manera integral a financiamiento para el agro"
recomendaciones = recomendar_instrumentos(consulta_usuario)

print("Instrumentos recomendados:")
for rec in recomendaciones:
    print(f"Nombre del Instrumento: {rec['Nombre del Instrumento']}")
    print(f"Descripción: {rec['Descripción']}")
    print(f"Explicación: {rec['Explicacion']}")
    print()
  

```

```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from langchain.chains import RetrievalQA
from langchain_groq import ChatGroq
from sklearn.metrics.pairwise import cosine_similarity
import getpass
import os
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

if not os.environ.get("GROQ_API_KEY"):
    os.environ["GROQ_API_KEY"] = getpass.getpass("Enter API key for Groq: ")

model = ChatGroq(
    model="llama-3.3-70b-versatile",
    max_tokens=6000  # Limitar tokens de salida
)

def recomendar_instrumentos(query, data_arco, qa_chain):
    """
    Recomienda instrumentos de política pública y genera un párrafo cohesivo.
    """
    resultados = qa_chain.run(query)
    instrumentos_recomendados = []

    tfidf = TfidfVectorizer()
    tfidf_matrix = tfidf.fit_transform(data_arco['Descripción'])

    # Limitar a 2 instrumentos más relevantes
    max_instrumentos = 2
    
    for descripcion in resultados.split("\n"):
        if len(instrumentos_recomendados) >= max_instrumentos:
            break
            
        descripcion = descripcion.strip()
        if descripcion:
            query_vector = tfidf.transform([descripcion])
            similitudes = cosine_similarity(query_vector, tfidf_matrix)
            indices_similares = similitudes.argsort()[0][-1:][::-1]  # Solo el más similar

            for idx in indices_similares:
                instrumento = {
                    "Nombre del Instrumento": data_arco.iloc[idx]['Nombre del Instrumento'],
                    "Descripción": data_arco.iloc[idx]['Descripción'][:300]  # Limitar longitud
                }
                instrumentos_recomendados.append(instrumento)
    
    # Generar prompt más conciso
    nombres_instrumentos = [i["Nombre del Instrumento"] for i in instrumentos_recomendados]
    prompt_resumen = f"Genera un párrafo breve que explique cómo estos instrumentos: {', '.join(nombres_instrumentos)} se complementan para atender la necesidad: {query}"
    
    parrafo_resumen = model.invoke(prompt_resumen).content
    
    return {
        "instrumentos": instrumentos_recomendados,
        "resumen": parrafo_resumen
    }

# Configuración y uso
#nombre_archivo = "2024 11 01 - Base Original.xlsx"
#data_arco = pd.read_excel(nombre_archivo)
data_arco = data_arco[['Nombre del Instrumento', 'Descripción']].dropna()

embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
vector_store = FAISS.from_texts(data_arco['Descripción'].tolist(), embeddings)
qa_chain = RetrievalQA.from_chain_type(llm=model, retriever=vector_store.as_retriever())

# Ejemplo de uso
consulta = "Dame una ruta de instrumentos para que un usuario acceda de manera integral a financiamiento para el agro"
resultado = recomendar_instrumentos(consulta, data_arco, qa_chain)

print("\nResumen de recomendaciones:")
print(resultado["resumen"])
print("\nInstrumentos recomendados:")
for instrumento in resultado["instrumentos"]:
    print(f"\nNombre: {instrumento['Nombre del Instrumento']}")
    print(f"Descripción: {instrumento['Descripción']}")
```
