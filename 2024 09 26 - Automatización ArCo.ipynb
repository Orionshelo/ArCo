## Selección de hojas de cálculo del Informe ArCo

```python
# Armado de la base -> Ruta de cada hoja, con los distintos tipos de información
general = pd.read_excel(nombre_archivo, sheet_name="1. Informe General")

beneficios = pd.read_excel(nombre_archivo, sheet_name="3. Beneficios")

Cobertura = pd.read_excel(nombre_archivo, sheet_name="4. Cobertura")

OPP = pd.read_excel(nombre_archivo, sheet_name="6. OPP")

Basica = pd.read_excel(nombre_archivo, sheet_name="7, Informacion Basica")

hash_tag = pd.read_excel(nombre_archivo, sheet_name="8, HashTag")

Dirigido_ampliado = pd.read_excel(nombre_archivo, sheet_name="9, Dirigido a")

test = pd.read_excel(nombre_archivo, sheet_name="10, Test")

Dirigido = pd.read_excel(nombre_archivo, sheet_name="11, Dirigido segun req.")

Beneficios_requerimiento = pd.read_excel(nombre_archivo, sheet_name="12, Beneficios Req.")

Localización = pd.read_excel(nombre_archivo, sheet_name="14, Localizacion")

Financiacion = pd.read_excel(nombre_archivo, sheet_name="15, Financiamiento Req.")

Fechas = pd.read_excel(nombre_archivo, sheet_name="16, Fechas")

descripciones_sep = pd.read_excel(nombre_archivo, sheet_name="17, Descripciones separadas")

Dirigido_ampliado = Dirigido_ampliado.drop(Dirigido_ampliado.columns[[4]], axis=1)
```

## Unión de bases

```python
# Unión de las distintas bases de datos
"""
se unen por medio del ContentID/Código para encontrar la coincidencia en
las distintas hojas del archivo de excel
"""

PR = pd.merge(general, Beneficios_requerimiento, on=["ContentID", "Entidad", "Nombre_Instrumento"], how="outer")
PR = pd.merge(PR, beneficios, on=["ContentID", "Entidad", "Nombre_Instrumento"], how="outer")
PR = pd.merge(PR, Dirigido, on=["ContentID", "Entidad", "Nombre_Instrumento"], how="outer")
PR = pd.merge(PR, Basica, on=["ContentID", "Entidad", "Nombre_Instrumento"], how="outer")
PR = pd.merge(PR, Cobertura, on=["ContentID", "Entidad", "Nombre_Instrumento"], how="outer")
PR = pd.merge(PR, Localización, on=["ContentID"], how="outer")
PR = pd.merge(PR, OPP, on=["ContentID", "Entidad", "Nombre_Instrumento"], how="outer")
PR = pd.merge(PR, test, on=["ContentID", "Entidad", "Nombre_Instrumento"], how="outer")
PR = pd.merge(PR, Dirigido_ampliado, on=["ContentID", "Entidad", "Nombre_Instrumento"], how="outer")
PR = pd.merge(PR, Financiacion, on=["ContentID", "Entidad", "Nombre_Instrumento"], how="outer")
PR = pd.merge(PR, Fechas, on=["ContentID", "Nombre_Instrumento"], how="outer")
PR = pd.merge(PR, descripciones_sep, on=["ContentID", "Nombre_Instrumento"], how="outer")
```

## Desagregación de Departamentos y Municipios

```python
# Importar las bibliotecas necesarias
import pandas as pd

# Paso 1: Cargar el archivo Excel
ruta_archivo = 'C:/Backup - Archivos de trabajo/DNP/Tareas/2024/Octubre/2024 10 18 - Power BI/Insumos/2024 10 21 - Base Original.xlsx'  # Cambia esta ruta por la ubicación de tu archivo
df = pd.read_excel(ruta_archivo)

# Paso 2: Visualizar el DataFrame original (opcional, para entender su estructura)
print(df.head())

# Paso 3: Crear una función para replicar filas con la desagregación, manejando valores nulos
def desagregar_columna(df, columna):
    # Dividir la columna por comas, manejando NaN
    desagregado = df[columna].fillna('').str.split(',')
    
    # Crear un nuevo DataFrame replicando las filas según la cantidad de elementos
    df_expanded = df.loc[df.index.repeat(desagregado.str.len())].copy()
    
    # Asignar los valores desagregados a la columna original sin filtrar los vacíos
    df_expanded[columna] = [elem.strip() for sublist in desagregado for elem in sublist]
    
    return df_expanded

# Paso 4: Desagregar la columna de 'Departamentos'
df_departamentos = desagregar_columna(df, 'Departamento')

# Paso 5: Desagregar la columna de 'Municipios'
df_municipios = desagregar_columna(df, 'Municipio')


# Paso 6: Exportar los resultados a archivos Excel separados
df_departamentos.to_excel('C:/Backup - Archivos de trabajo/DNP/Tareas/2024/Octubre/2024 10 18 - Power BI/Insumos/2021 10 21 - departamentos_desagregados.xlsx', index=False)
df_municipios.to_excel('C:/Backup - Archivos de trabajo/DNP/Tareas/2024/Octubre/2024 10 18 - Power BI/Insumos/2021 10 21 - municipios_desagregados.xlsx', index=False)
# Mensaje de éxito
print("Los archivos 'departamentos_desagregados.xlsx' y 'municipios_desagregados.xlsx' han sido generados exitosamente.")
```

```python
import pandas as pd
import numpy as np
from openpyxl import Workbook
from openpyxl.utils.dataframe import dataframe_to_rows
import os

# Listas proporcionadas
tipos_apoyo = ['Formación de Talento Humano',
               'Tipo de Sistemas de Información',
               'Asistencia técnica y Consultoría',
               'Tipo de Redes de Colaboración',
               'Apoyo Financiero',
               'Incentivos Tributarios',
               'Eventos',
               'Compra Pública',
               'Bonos o Vouchers',
               'Premios y Reconocimientos',
               'Instrumentos Regulatorios']

objetivos_politica = ["Innovación", "Emprendimiento", "Economía Popular",
                     "Transferencia de Conocimiento y Tecnología", "Investigación",
                     "Formación de Capital Humano", "Propiedad intelectual", "Calidad",
                     "Clúster o encadenamientos", "Financiación", "Comercialización",
                     "Formalización", "Crecimiento Sostenible", "Inclusión Financiera",
                     "Comercio Electrónico", "Educación económica y financiera", "Competencias Digitales"]

tipos_usuario = ["Emprendedores",
                "Mipymes",
                "Grandes empresas",
                "Gobierno ",
                "Academia",
                "Entidades de soporte",
                "Personas naturales"]


def cargar_excel(ruta_archivo):
    df = pd.read_excel(ruta_archivo)
    
    # Verificar y eliminar filas completamente vacías
    df = df.dropna(how='all')
    
    # Verificar duplicados exactos y eliminarlos si existen
    duplicados = df.duplicated()
    if duplicados.any():
        print(f"Se encontraron {duplicados.sum()} filas duplicadas exactas")
        df = df.drop_duplicates()
    
    # Imprimir información de diagnóstico
    print(f"Total de filas en el archivo: {len(df)}")
    print(f"Total de sectores únicos: {df['Sector'].nunique()}")
    
    return df

def procesar_datos(df):
    # Crear una copia del DataFrame para no modificar el original
    df_processed = df.copy()
    
    # Convert 'Si' to 1 and 'No' to 0 for all relevant columns
    for columna in objetivos_politica + tipos_apoyo + tipos_usuario:
        if columna in df_processed.columns:
            df_processed[columna] = np.where(df_processed[columna].str.strip().str.lower() == 'si', 1, 0)
    
    # Calculate global metrics
    total_instrumentos = len(df_processed)
    print(f"Total de instrumentos después del procesamiento: {total_instrumentos}")
    
    # Calculate metrics for objectives
    objetivos_por_instrumento = df_processed[objetivos_politica].sum(axis=1)
    porcentaje_un_objetivo = (objetivos_por_instrumento == 1).mean() * 100
    
    # Calculate metrics for users
    usuarios_por_instrumento = df_processed[tipos_usuario].sum(axis=1)
    porcentaje_pocos_usuarios = ((usuarios_por_instrumento == 1) | (usuarios_por_instrumento == 2)).mean() * 100
    
    # Calculate metrics for support types
    tipos_apoyo_por_instrumento = df_processed[tipos_apoyo].sum(axis=1)
    porcentaje_un_apoyo = (tipos_apoyo_por_instrumento == 1).mean() * 100
    
    # Print diagnostic information
    print("\nDistribución de objetivos por instrumento:")
    print(objetivos_por_instrumento.value_counts().sort_index())
    print("\nDistribución de usuarios por instrumento:")
    print(usuarios_por_instrumento.value_counts().sort_index())
    print("\nDistribución de tipos de apoyo por instrumento:")
    print(tipos_apoyo_por_instrumento.value_counts().sort_index())
    
    resultados_globales = {
        'Total de instrumentos': total_instrumentos,
        'Mejor enfoque (un objetivo)': porcentaje_un_objetivo,
        'Mayor focalización (1-2 usuarios)': porcentaje_pocos_usuarios,
        'Mejor identificación (un tipo apoyo)': porcentaje_un_apoyo
    }
    
    # Process sector-specific analysis
    resultados_sectores = df_processed.groupby('Sector').apply(analizar_sector)
    
    # Verificar que la suma de instrumentos por sector coincida con el total
    total_por_sectores = resultados_sectores['Total Instrumentos'].sum()
    print(f"\nTotal de instrumentos sumando por sectores: {total_por_sectores}")
    if total_por_sectores != total_instrumentos:
        print("¡ADVERTENCIA! La suma de instrumentos por sector no coincide con el total global")
    
    return resultados_globales, resultados_sectores

def analizar_sector(sector_data):
    # El resto de la función se mantiene igual
    total_instrumentos_sector = len(sector_data)
    
    objetivos_por_instrumento = sector_data[objetivos_politica].sum(axis=1)
    porcentaje_un_objetivo = (objetivos_por_instrumento == 1).mean() * 100
    
    usuarios_por_instrumento = sector_data[tipos_usuario].sum(axis=1)
    #porcentaje_pocos_usuarios = ((usuarios_por_instrumento == 1) | (usuarios_por_instrumento == 2)).mean() * 100
    porcentaje_pocos_usuarios = ((usuarios_por_instrumento < 3)).sum()

    
    tipos_apoyo_por_instrumento = sector_data[tipos_apoyo].sum(axis=1)
    porcentaje_un_apoyo = (tipos_apoyo_por_instrumento == 1).mean() * 100
    
    objetivos_count = sector_data[objetivos_politica].sum()
    apoyos_count = sector_data[tipos_apoyo].sum()
    usuarios_count = sector_data[tipos_usuario].sum()
    
    return pd.Series({
        'Mejor enfoque': porcentaje_un_objetivo,
        'Mayor focalización': porcentaje_pocos_usuarios,
        'Mejor identificación': porcentaje_un_apoyo,
        'Total Instrumentos': total_instrumentos_sector,
        'Objetivos de Política': objetivos_count.to_dict(),
        'Tipos de Apoyo': apoyos_count.to_dict(),
        'Tipos de Usuario': usuarios_count.to_dict()
    })

def generar_excel_resultados(resultados_globales, resultados_sectores, ruta_salida):
    # La función se mantiene igual que en la versión anterior
    wb = Workbook()
    
    # Create Global Results sheet
    ws_global = wb.active
    ws_global.title = "Resultados Globales"
    
    ws_global.append(['Indicador', 'Valor'])
    for metrica, valor in resultados_globales.items():
        if isinstance(valor, (int, np.integer)):
            ws_global.append([metrica, valor])
        else:
            ws_global.append([metrica, f"{valor:.1f}%"])
    
    # Create Sector Results sheets
    for sector in resultados_sectores.index:
        ws_sector = wb.create_sheet(title=sector[:31])
        
        ws_sector.append(['Indicador', 'Valor'])
        ws_sector.append(['Mejor enfoque', f"{resultados_sectores.loc[sector, 'Mejor enfoque']:.1f}%"])
        ws_sector.append(['Mayor focalización', f"{resultados_sectores.loc[sector, 'Mayor focalización']:.1f}%"])
        ws_sector.append(['Mejor identificación', f"{resultados_sectores.loc[sector, 'Mejor identificación']:.1f}%"])
        ws_sector.append(['Total Instrumentos', resultados_sectores.loc[sector, 'Total Instrumentos']])
        
        ws_sector.append([])
        ws_sector.append(['Objetivos de Política', 'Conteo'])
        for objetivo, conteo in resultados_sectores.loc[sector, 'Objetivos de Política'].items():
            ws_sector.append([objetivo, conteo])
        
        ws_sector.append([])
        ws_sector.append(['Tipos de Apoyo', 'Conteo'])
        for apoyo, conteo in resultados_sectores.loc[sector, 'Tipos de Apoyo'].items():
            ws_sector.append([apoyo, conteo])
            
        ws_sector.append([])
        ws_sector.append(['Tipos de Usuario', 'Conteo'])
        for usuario, conteo in resultados_sectores.loc[sector, 'Tipos de Usuario'].items():
            ws_sector.append([usuario, conteo])
    
    wb.save(ruta_salida)

def main(ruta_entrada, ruta_salida):
    if not os.path.exists(ruta_entrada):
        print(f"Error: El archivo de entrada '{ruta_entrada}' no existe.")
        return

    directorio_salida = os.path.dirname(ruta_salida)
    if not os.path.exists(directorio_salida):
        os.makedirs(directorio_salida)
        print(f"Se ha creado el directorio: {directorio_salida}")

    try:
        print("Iniciando procesamiento del archivo...")
        df = cargar_excel(ruta_entrada)
        resultados_globales, resultados_sectores = procesar_datos(df)
        generar_excel_resultados(resultados_globales, resultados_sectores, ruta_salida)
        print(f"Informe generado y guardado en {ruta_salida}")
    except Exception as e:
        print(f"Error al procesar el archivo: {str(e)}")

# Script usage
ruta_entrada = "C:/Users/supegui/Documents/Metodología ArCo/Python/Excel/2024 11 01 - Base Original.xlsx"
ruta_salida = "C:/Users/supegui/Documents/Metodología ArCo/Python/Excel/Resultados 1era Iteración 2025/analisis_1.xlsx"
main(ruta_entrada, ruta_salida)
```
