```python
import re

# Function to remove phrases
def remove_phrases(text):
    for phrase in phrases_to_exclude:
        # Create a regex pattern for the phrase with optional punctuation
        pattern = re.escape(phrase) + r'[\.,;"]*'
        text = re.sub(pattern, '', text)
    return text.strip()

# Apply the function to the 'Descripción larga' column
arco['Descripción larga'] = arco['Descripción larga'].apply(remove_phrases)

# Display the updated descriptions
print(arco['Descripción larga'].head())
```

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from difflib import SequenceMatcher

# Download NLTK stopwords if not already downloaded
nltk.download('stopwords')

# Define Spanish stopwords
spanish_stopwords = set(stopwords.words('spanish'))

spanish_stopwords = list(stopwords.words('spanish'))

# Load your datasets
# Replace 'agenda_nacional_ci' and 'arco' with your actual dataframes
# agenda_nacional_ci = pd.read_csv('agenda_nacional_ci.csv')
# arco = pd.read_csv('arco.csv')

# Extract relevant columns
alcances = agenda_nacional_ci['Alcances'].dropna().astype(str)
eje_comp_inno = agenda_nacional_ci['Eje para la Competitividad e Innovación'].dropna().astype(str)
acciones_proyectos = agenda_nacional_ci['Acciones/Proyectos'].dropna().astype(str)
descripciones_largas = arco['Descripción larga'].dropna().astype(str)
nombres_instrumentos = arco['Nombre del Instrumento'].dropna().astype(str)
```

```python
# Initialize TfidfVectorizer with Spanish stopwords and n-grams
vectorizer = TfidfVectorizer(stop_words=spanish_stopwords, ngram_range=(1, 2))

# Combine 'acciones_proyectos' and 'descripciones_largas' for vectorization
combined_texts = pd.concat([acciones_proyectos, descripciones_largas], ignore_index=True)

# Fit and transform the vectorizer
tfidf_matrix = vectorizer.fit_transform(combined_texts)
```

```python
# Split the TF-IDF matrix into two parts
tfidf_acciones = tfidf_matrix[:len(acciones_proyectos)]
tfidf_descripciones = tfidf_matrix[len(acciones_proyectos):]

# Compute cosine similarity matrix
similarity_matrix = cosine_similarity(tfidf_acciones, tfidf_descripciones)
```

```python
# Get indices of the top matches
top_match_indices = similarity_matrix.argmax(axis=1)

# Get the maximum similarity scores
max_similarities = similarity_matrix.max(axis=1)
```

```python
# Function to find the best match using difflib
def find_best_match(action, descriptions):
    best_match = ''
    best_ratio = 0
    for desc in descriptions:
        ratio = SequenceMatcher(None, action, desc).ratio()
        if ratio > best_ratio:
            best_ratio = ratio
            best_match = desc
    return best_match, best_ratio

# Lists to store the results
best_matches = []
best_ratios = []

# Find the best match for each 'Acciones/Proyectos'
for action in acciones_proyectos:
    match, ratio = find_best_match(action, descripciones_largas)
    best_matches.append(match)
    best_ratios.append(ratio)
```

```python
# Create a DataFrame to store the results
resultados_similitud = pd.DataFrame({
    'Acciones/Proyectos': acciones_proyectos,
    'Descripción larga más similar (TF-IDF)': [descripciones_largas.iloc[idx] for idx in top_match_indices],
    'Similitud TF-IDF': max_similarities,
    'Descripción larga más similar (difflib)': best_matches,
    'Similitud difflib': best_ratios,
    'Tiene similitud': ['Sí' if sim > 0.7 else 'No' for sim in max_similarities]  # Adjust threshold as needed
})

# Optionally, include 'Alcances' and 'Eje para la Competitividad e Innovación'
resultados_similitud['Alcances'] = alcances.reset_index(drop=True)
resultados_similitud['Eje para la Competitividad e Innovación'] = eje_comp_inno.reset_index(drop=True)
resultados_similitud['Nombre del Instrumento más similar'] = [nombres_instrumentos.iloc[idx] for idx in top_match_indices]
```

```python
import numpy as np
import pandas as pd
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Download NLTK stopwords if not already downloaded
nltk.download('stopwords')

# Define Spanish stopwords as a list
spanish_stopwords = list(stopwords.words('spanish'))

# Load your datasets
# Replace 'agenda_nacional_ci' and 'arco' with your actual dataframes
# agenda_nacional_ci = pd.read_csv('agenda_nacional_ci.csv')
# arco = pd.read_csv('arco.csv')

# Extract relevant columns and ensure they are strings without missing values
acciones_proyectos = agenda_nacional_ci['Acciones/Proyectos'].dropna().astype(str)
descripciones_largas = arco['Descripción larga'].dropna().astype(str)
nombres_instrumentos = arco['Nombre del Instrumento'].dropna().astype(str)

# Combine the texts for vectorization with consistent indexing
combined_texts = pd.concat([acciones_proyectos.reset_index(drop=True), descripciones_largas.reset_index(drop=True)], ignore_index=True)

# Initialize TfidfVectorizer with Spanish stopwords and n-grams
vectorizer = TfidfVectorizer(stop_words=spanish_stopwords, ngram_range=(1, 2))

# Fit and transform the vectorizer
tfidf_matrix = vectorizer.fit_transform(combined_texts)

# Split the TF-IDF matrix into two parts
tfidf_acciones = tfidf_matrix[:len(acciones_proyectos)]
tfidf_descripciones = tfidf_matrix[len(acciones_proyectos):]

# Compute cosine similarity matrix
similarity_matrix = cosine_similarity(tfidf_acciones, tfidf_descripciones)

# Define the number of top matches to retrieve
N = 3  # You can adjust this number based on your needs

# Get indices of the top N matches for each project
top_match_indices = np.argsort(similarity_matrix, axis=1)[:, -N:]

# Get the top N similarity scores for each project
top_similarities = np.sort(similarity_matrix, axis=1)[:, -N:]

# Function to get the top N instrument descriptions and similarities
def get_top_n_descriptions(indices, similarities, instrument_names, descriptions):
    descriptions_list = []
    for idx, sim in zip(indices, similarities):
        desc = descriptions.iloc[idx]
        descriptions_list.append(f"{instrument_names.iloc[idx]} - Similarity: {sim:.2f}\n{desc}")
    return descriptions_list

# List to store the top N descriptions for each project
top_n_descriptions = []

for i in range(len(acciones_proyectos)):
    indices = top_match_indices[i][::-1]  # Reverse to get highest first
    sims = top_similarities[i][::-1]
    descriptions = get_top_n_descriptions(indices, sims, nombres_instrumentos, descripciones_largas)
    top_n_descriptions.append(descriptions)

# Create a DataFrame to store the results
resultados_similitud = pd.DataFrame({
    'Acciones/Proyectos': acciones_proyectos,
    'Top N Instrumentos más similares': [', '.join([f"{nombres_instrumentos.iloc[idx]} (Similarity: {sim:.2f})" for idx, sim in zip(top_match_indices[i][::-1], top_similarities[i][::-1])]) for i in range(len(acciones_proyectos))]
})

# Create new columns for each of the top N descriptions
for n in range(N):
    resultados_similitud[f'Top {n+1} Instrumento - Descripción'] = [desc[n] if n < len(desc) else '' for desc in top_n_descriptions]

# Function to concatenate descriptions into a single string
def concatenate_descriptions(descriptions_list):
    return '\n\n'.join(descriptions_list)

# Add a column with all top N descriptions concatenated
resultados_similitud['Top N Instrumentos - Descripciones Concatenadas'] = [concatenate_descriptions(desc) for desc in top_n_descriptions]

# Display the updated results
print(resultados_similitud.head())
```
